Now you can use the `translator` pipeline to translate English sentences to French. Simply pass the English sentence as a string to the `translator` object.

```python
translator = pipeline("translation_en_to_fr")
```

```python
english_sentence = "good morning"
french_translation = translator(english_sentence)
print(french_translation)
```
